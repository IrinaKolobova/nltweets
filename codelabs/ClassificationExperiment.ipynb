```python
labels = []
texts = []
with open("labeled_tweets.csv", "r") as f:
    for line in f.readlines():
        arr = line.split(",")
        if arr[0] == "Topic":
            continue
        label = arr[0]
        rest = arr[1:]
        text = ",".join(rest)
        labels.append(int(label))
        texts.append(text)
```

```python
from sklearn.utils import shuffle
texts, labels = shuffle(texts, labels)
train_texts = texts[:600]
train_labels = labels[:600]
test_texts = texts[600:]
test_labels = labels[600:]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
X = vectorizer.fit_transform(train_texts)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, max_depth=2,random_state=0)
```

```python
clf.fit(X, train_labels)
```

```python
X_test = vectorizer.transform(test_texts)
res = clf.predict(X_test)
res
```

```python
test_labels
```

```python
print(sum([1 if test_labels[i]-resi==0 else 0 for (i, resi) in enumerate(res)])/len(res))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm
for clf in ([
        RandomForestClassifier(),
        svm.SVC(class_weight={1: 4.08}),
        LogisticRegression(C=1000),
    ]):
    clf.fit(X, train_labels)
    X_test = vectorizer.transform(test_texts)
    res = clf.predict(X_test)
    print(res)
    print(sum([1 if test_labels[i]-resi==0 else 0 for (i, resi) in enumerate(res)])/len(res))
```
